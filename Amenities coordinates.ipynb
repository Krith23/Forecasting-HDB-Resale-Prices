```python
# Function to retrieve latitude and longitude for an address
def get_lat_long(address):
    url = f"https://www.onemap.gov.sg/api/common/elastic/search?searchVal={address}&returnGeom=Y&getAddrDetails=Y&pageNum=1"
    response = requests.get(url)
    data = response.json()
    if data['found'] > 0:
        latitude = data['results'][0]['LATITUDE']
        longitude = data['results'][0]['LONGITUDE']
        return latitude, longitude
    else:
        return None, None
```

```python
tqdm.pandas(desc="Progress")
shops_df[['latitude', 'longitude']] = shops_df['trimmed_address'].progress_apply(lambda x: pd.Series(get_lat_long(x)))
```

```python
# Function to retrieve latitude and longitude for an address
def get_lat_long_name(station_code):
    url = f"https://www.onemap.gov.sg/api/common/elastic/search?searchVal={station_code}&returnGeom=Y&getAddrDetails=Y&pageNum=1"
    response = requests.get(url)
    data = response.json()
    if data['found'] > 0:
        latitude = data['results'][0]['LATITUDE']
        longitude = data['results'][0]['LONGITUDE']
        station_name = data['results'][0]['SEARCHVAL']
        return latitude, longitude, station_name
    else:
        return None, None


# Add latitude and longitude columns to the address DataFrame
tqdm.pandas(desc="Progress")
mrt_df[['latitude', 'longitude', 'station_name']] = mrt_df['station_code'].progress_apply(lambda x: pd.Series(get_lat_long_name(x)))
```
