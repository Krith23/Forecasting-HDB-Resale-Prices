```python
import pandas as pd
import requests
from tqdm import tqdm

# Function to retrieve latitude and longitude for an address
def get_lat_long(address):
    url = f"https://www.onemap.gov.sg/api/common/elastic/search?searchVal={address}&returnGeom=Y&getAddrDetails=Y&pageNum=1"
    response = requests.get(url)
    data = response.json()
    if data['found'] > 0:
        latitude = data['results'][0]['LATITUDE']
        longitude = data['results'][0]['LONGITUDE']
        return latitude, longitude
    else:
        return None, None


# Add latitude and longitude columns to the address DataFrame
tqdm.pandas(desc="Progress")
address_df[['latitude', 'longitude']] = address_df['address'].progress_apply(lambda x: pd.Series(get_lat_long(x)))
```

```python
# Merge the coordinates with hdb_df based on the 'Address' column
hdb_df = hdb_df.merge(address_df, on='address', how='left')
```

```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import MarkerCluster
import math

# Create the map centered around a general location (adjust the coordinates as needed)
m_3 = folium.Map(location=[1.3521, 103.8198], tiles='cartodbpositron', zoom_start=13)

# Add points to the map
mc = MarkerCluster()
for idx, row in address_df.iterrows():
    if not math.isnan(row['longitude']) and not math.isnan(row['latitude']):
        mc.add_child(Marker([row['latitude'], row['longitude']]))
        
m_3.add_child(mc)

# Display the map
m_3
```

```python
# Convert 'month' column to datetime and extract the year
hdb_df['year'] = pd.to_datetime(hdb_df['month']).dt.year

# Filter the DataFrame to include only records from the year 2023
hdb_df_2023 = hdb_df[hdb_df['year'] == 2023]

# Calculate average price for each address, latitude, and longitude
avg_price_df = hdb_df_2023.groupby(['address', 'latitude', 'longitude'])['resale_price'].mean().reset_index()
```
